# 6.결측값을 다른 값으로 대체

```python
nba = pd.read_csv("02_nba.csv").dropna(how = "all")
nba.head()
```

```python
nba["Salary"].fillna(0, inplace = True)
nba["College"].fillna("None", inplace = True)
nba.head()
```

```python
nba.dtypes
```

```python
nba['Salary'].head()
```

```python
nba['Salary'].astype('int').head()
```

```python
nba['Position'].nunique() # 동일한 숫자가 몇개인지 확인
```

```python
nba['Position'] = nba['Position'].astype('category')
nba['Position'].head()
```

```python
nba['Team'].nunique()
```

# 7. 정렬

```python
nba = pd.read_csv("02_nba.csv")
nba.head()
```

```python
nba.sort_values("Name", ascending = False).head() # 내림차순
```

```python
nba.sort_values(['Team', 'Name'], ascending=False).head() # 두 개의 컬럼을 사용해서 내림 차순
```

```python
nba.sort_values(['Team', 'Name'], ascending=[True, False], inplace=True) # 두 개의 컬럼을 개별로 정렬 기준을 설정하며, 내용을 반영함
nba.head()
```

```python
nba.sort_index(ascending=False, inplace=True) # 인덱스를 기준으로 정렬
nba.head()
```

```python
nba = pd.read_csv("02_nba.csv").dropna(how = "all")
nba['Salary Rank'] = nba['Salary'].fillna(0).astype('int')
nba.head()
```

```python
nba['Salary Rank'].rank(ascending=False).astype('int') # Salary Rank를 기준으로 정렬(rank)
```

```python
nba.sort_values(by = 'Salary', ascending = False).head()
```
