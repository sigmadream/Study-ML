# Filtering with String Methods

```python
chicago = pd.read_csv("06_chicago.csv").dropna(how = "all")
chicago["Department"] = chicago["Department"].astype("category")
chicago.tail(3)
```

```python
mask = chicago["Position Title"].str.lower().str.contains("water")
chicago[mask]
```

```python
chicago[chicago["Position Title"].str.lower().str.startswith("water")]
```

```python
mask = chicago["Position Title"].str.lower().str.endswith("ist")
chicago[mask]
```

# More String Methods - .strip(), .lstrip(), and .rstrip()

```python
chicago = pd.read_csv("06_chicago.csv").dropna(how = "all")
chicago["Department"] = chicago["Department"].astype("category")
chicago.tail(3)
```

```python
"               Hello World  ".rstrip()
```

```python
"               Hello World  ".strip()
```

```python
chicago["Name"] = chicago["Name"].str.rstrip().str.lstrip()
chicago["Name"]
```

```python
chicago["Position Title"] = chicago["Position Title"].str.strip()
chicago["Position Title"]
```

# String Methods on Index and Columns

# The expand and n Parameters of the str.split() Method

```python
chicago = pd.read_csv("06_chicago.csv").dropna(how = "all")
chicago["Department"] = chicago["Department"].astype("category")
chicago.tail(3)
```

```python
chicago[["First Name", "Last Name"]] = chicago["Name"].str.split(",", expand = True)
```

```python
chicago.head(3)
```

```python
chicago[["First Title World", "Remainign Words"]] = chicago["Position Title"].str.split(" ", expand = True, n = 1)
```
